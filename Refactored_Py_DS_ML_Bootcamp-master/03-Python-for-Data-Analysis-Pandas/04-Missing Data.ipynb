# Missing Data

Let's show a few convenient methods to deal with Missing Data in pandas:

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3]})
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(thresh=2)
```

```python
df2 = pd.DataFrame(np.random.randn(10,10),index=["A","B","C","D","E","F","G","H","I","J"],columns=["One","Two","Three","Four","Five","Six","Seven","Eight","Nine","Ten"])
df2
```

```python

df2 = df2[df2>0]
df2
```

```python
df2.fillna(value="Fix")

```

```python
df2["One"].fillna(value=df2["One"].mean())
```

```python
df.fillna(value='FILL VALUE')
```

```python
df['A'].fillna(value=df['A'].mean())
```
