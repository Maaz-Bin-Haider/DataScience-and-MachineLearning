# NumPy 

NumPy (or Numpy) is a Linear Algebra Library for Python, the reason it is so important for Data Science with Python is that almost all of the libraries in the PyData Ecosystem rely on NumPy as one of their main building blocks.

Numpy is also incredibly fast, as it has bindings to C libraries. For more info on why you would want to use Arrays instead of lists, check out this great [StackOverflow post](http://stackoverflow.com/questions/993984/why-numpy-instead-of-python-lists).

We will only learn the basics of NumPy, to get started we need to install it!

# Numpy Arrays

NumPy arrays are the main way we will use Numpy throughout the course. Numpy arrays essentially come in two flavors: vectors and matrices. Vectors are strictly 1-d arrays and matrices are 2-d (but you should note a matrix can still have only one row or one column).

Let's begin our introduction by exploring how to create NumPy arrays.

## Creating NumPy Arrays

```python
np.array(my_list)
```

```python
my_matrix = [[1,2,3],[4,5,6],[7,8,9]]
my_matrix
```

### linspace
Return 1-D Array evenly spaced numbers over a specified interval.



linspace is a NumPy function used to generate evenly spaced numbers over a specified interval.

numpy.linspace(start, stop, num=50, endpoint=True)

Parameters Explained
1. start

The starting value of the sequence

2. stop

The ending value of the sequence

3. num (optional)

Number of values to generate

Default is 50

4. endpoint (optional)

If True → includes the stop value

If False → excludes the stop value

```python
np.linspace(0,5,50,endpoint=True)
```

```python
np.linspace(10,20,20,endpoint=False)
```

## eye

Creates an identity matrix

## What does the eye() function do in NumPy?

The eye() function in NumPy is used to create an identity matrix .

np.eye(N, M=None, k=0)

### Parameters Explained
1. N

Number of rows

2. M (optional)

Number of columns

If not given → matrix will be square (N × N)

3. k (optional)

Position of the diagonal

k = 0 → main diagonal (default)

k > 0 → above main diagonal

k < 0 → below main diagonal

```python
np.eye(4)
```

```python
np.eye(3,4) # an square Identity matrix
```

```python
np.eye(6,4)
```

```python
# k = 0 → main diagonal (default)
np.eye(4,k=0)
```

```python
# k > 0 → above main diagonal
np.eye(4,k=1)
```

```python
# k < 0 → below main diagonal
np.eye(4,k=-1)
```

## Random 

Numpy also has lots of ways to create random number arrays:

## rand
Create an array of the given shape and populate it with
random samples from a uniform distribution
over ``[0, 1)``.

### 0 ≤ random number < 1

```python
np.random.rand(2)
```

```python
arr = np.random.rand(5,5)
arr
```

## randn

Return a sample (or samples) from the "standard normal" distribution. Unlike rand which is uniform:

### What does randn() do in NumPy?

np.random.randn() generates random numbers from a normal (Gaussian) distribution.

📌 This distribution has:

### Mean (μ) = 0

### Standard deviation (σ) = 1

So values can be:

Positive or negative

Small or large (but most are near 0)

```python
np.random.randn(5)
```

```python
np.random.randn(5,5)
```

## randint
Return random integers from `low` (inclusive) to `high` (exclusive).

### What does randint() do in NumPy?

np.random.randint() generates random integers between a given range.

Values are whole numbers

Lower bound is inclusive

Upper bound is exclusive

``np.random.randint(low, high=None, size=None)``

### Parameters Explained
1️⃣ low

Minimum value (included)

2️⃣ high (optional)

Maximum value (excluded)

If not given → range is 0 to low

3️⃣ size (optional)

Shape of the output (single value, 1D, 2D, etc.)

## 2️⃣ 1D Array

```python
np.random.randint(1,100)
```

```python
np.random.randint(1,100,10)
```

## 2️⃣ 2D Array

```python
np.random.randint(1,100,size=(3,3))
```
